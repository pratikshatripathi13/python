```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks

    def get_marks(self):
        return self.__marks


s = Student("Pratiksha", 92)
print(s.name)
print(s.get_marks())
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Deposited:", amount)

    def check_balance(self):
        return self.__balance


a = BankAccount(1000)
a.deposit(500)
print(a.check_balance())
```

```python
class person:
    def __init__(self,age):
        self.__age=age
    def show(self):
        if(self.__age<0):
            print("invalid age")
        else:
            print(self.__age)
c=person(12)
c.show()
    
```

```python
class User:
    def __init__(self, password):
        self.__password = password

    def verify(self, pwd):
        if pwd == self.__password:
            print("Access granted")
        else:
            print("Wrong password")


u = User("abc123")
u.verify("abc123")
```

```python
class employee:
    def __init__(self,salary):
        self.__salary=salary
    def increase(self):
        self.__salary=self.__salary+1000
        return self.__salary
    
c=employee(4999)
c.increase()
```

```python
class Car:
    def __init__(self):
        self.__speed = 0

    def accelerate(self):
        self.__speed += 10

    def brake(self):
        self.__speed -= 5
        if self.__speed < 0:
            self.__speed = 0

    def show_speed(self):
        return self.__speed


c = Car()
c.accelerate()
c.brake()
print(c.show_speed())
```

```python
class Mobile:
    def __init__(self):
        self.__battery = 100

    def use(self):
        self.__battery -= 10
        if self.__battery < 0:
            self.__battery = 0

    def battery_status(self):
        return self.__battery


m = Mobile()
m.use()
print(m.battery_status())
```

```python
class ATM:
    def __init__(self, pin):
        self.__pin = pin

    def check_pin(self, entered):
        return entered == self.__pin


atm = ATM(1234)
print(atm.check_pin(1234))
```

```python
class book:
    def __init__(self,price):
        self.__price=price
    def check(self):
        if(self.__price<0):
            print("can't be neagtive")
        else:
            print(self.__price)
c=book(336)
c.check()
    
```

```python
class Result:
    def __init__(self):
        self.__marks = []

    def add(self, m):
        self.__marks.append(m)

    def total(self):
        return sum(self.__marks)


r = Result()
r.add(80)
r.add(90)
print(r.total())
```
