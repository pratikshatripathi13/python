```python
from abc import ABC, abstractmethod

class Course(ABC):

    @abstractmethod
    def duration(self):
        pass


class PythonCourse(Course):

    def duration(self):
        print("Python course duration: 3 months")


c = PythonCourse()
c.duration()
```

```python
from abc import ABC, abstractmethod

class Food(ABC):

    @abstractmethod
    def price(self):
        pass


class Pizza(Food):

    def price(self):
        print("Pizza price: ₹299")


f = Pizza()
f.price()
```

```python
from abc import ABC, abstractmethod
class Electronics(ABC):

    @abstractmethod
    def warranty(self):
        pass

class Laptop(Electronics):
    def warranty(self):
        print("Laptop warranty: 2 years")


e = Laptop()
e.warranty()
```
